#### Displaying a Stacked Bar Chart
Question 15 :
Write code to display a stacked bar chart of the count of species by island.

```python
df = pd.read_csv('penguins.csv')
tbl = pd.pivot_table(df.groupby(['island', 'species']).size().reset_index(), 
                     values=0,
                    index='island',
                    columns=['species'], 
                    aggfunc=np.sum)
tbl.plot(kind='bar', stacked=True)
```
